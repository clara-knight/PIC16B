```python
from square import square # import the square function we wrote

xs = [i for i in range(1, 11)]
xs_squared = [square(i) for i in xs]

print("List of x values:", xs)
print("List of x^2 values:", xs_squared)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1)
ax.scatter(x=xs, y=xs_squared, color='black')
ax.set(xlabel='x', ylabel='y')
fig.suptitle('$y = x^2$')
plt.show()
```
